## Ejercicio 1 

Dos dados se lanzan una vez y se observa el total obtenido. Usa una simulación para encontrar la probabilidad estimada de que la puntuación sea mayor a 7 o un número par. Una simulación es una repetición del mismo experimento multiples veces para observar su comportamiento:

- Ejecuta el experimento 1000 veces (lanza 2 dados 1000 veces, y suma el número de ambos dados).
- Lleva ua cuenta de los números y las veces que la suma fue mayor a 7 o un número par.
- Divide el número del paso 2 entre el número de iteraciones(1000).

```python
import random

def lanzar_dados():
    dado1 = random.randint(1, 6)  # Número aleatorio entre 1 y 6
    dado2 = random.randint(1, 6)
    return dado1 + dado2

def simulacion(iteraciones):
    conteo_exitos = 0
    
    for _ in range(iteraciones):
        suma = lanzar_dados()
        if suma > 7 or suma % 2 == 0:
            conteo_exitos += 1

    # Calcular la probabilidad estimada
    probabilidad = conteo_exitos / iteraciones
    return probabilidad

# Ejecutar la simulación con 1000 iteraciones
iteraciones = 1000
resultado = simulacion(iteraciones)

print(f"Probabilidad estimada de que la puntuación sea mayor a 7 o par: {resultado}")
```

## Ejercicio 2

Una caja contiene 10 bolas blancas, 20 bolas rojas y 30 bolas verdes. Si tomamos 5 bolas de la caja reemplazándolas (tomando una bola, anotando el color y luego regresandola a la caja). Queremos saber la probabilidad de:

1. Tomar 3 blancas y 2 rojas.
2. Toas todas del mismo color.

Ejecuta el experimento 1000 veces y calcula las probabilidades que mencionadas.

```python
import random

# Definir las bolas en la caja
bolas = ['blanca'] * 10 + ['roja'] * 20 + ['verde'] * 30

def tomar_bolas():
    #Función para simular la toma de 5 bolas.
    bolas_seleccionadas = random.choices(bolas, k=5)
    return bolas_seleccionadas

def simulacion(iteraciones):
    exito_3blancas_2rojas = 0
    exito_todas_mismo_color = 0
    
    for _ in range(iteraciones):
        bolas_seleccionadas = tomar_bolas()
        
        # Contar cuántas veces tomamos 3 bolas blancas y 2 bolas rojas
        blancas = bolas_seleccionadas.count('blanca')
        rojas = bolas_seleccionadas.count('roja')
        
        if blancas == 3 and rojas == 2:
            exito_3blancas_2rojas += 1
        
        # Contar cuántas veces todas las bolas son del mismo color
        if len(set(bolas_seleccionadas)) == 1:
            exito_todas_mismo_color += 1
    
    # Calcular las probabilidades
    prob_3blancas_2rojas = exito_3blancas_2rojas / iteraciones
    prob_todas_mismo_color = exito_todas_mismo_color / iteraciones
    
    return prob_3blancas_2rojas, prob_todas_mismo_color

# Ejecutar la simulación con 1000 iteraciones
iteraciones = 1000
prob_3blancas_2rojas, prob_todas_mismo_color = simulacion(iteraciones)

print(f"Probabilidad de tomar 3 blancas y 2 rojas: {prob_3blancas_2rojas}")
print(f"Probabilidad de que todas las bolas sean del mismo color: {prob_todas_mismo_color}")
```
